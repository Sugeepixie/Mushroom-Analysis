```python
df_f = pd.read_sql('select * from mushroom_features', con)
df_f
```

```python
df_t = pd.read_sql('select * from mushroom_targets', con)
df_t
```

```python
# Define the features and target

X = df_f.drop(["id"], axis=1)

y = df_t['class']
```

```python
for col in X.columns:
    X[col] = LabelEncoder().fit_transform(X[col])
```

```python
X
```

```python
# Split into training and testing sets
## Stratify?
## Check what random state should be set to

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=78)
```

```python
# Create a standard scaler instance
# scaler = StandardScaler()

# # Fit the Standard Scaler with the training data
# X_scaler = scaler.fit(X_train)

# # Scale the data
# X_train_scaled = X_scaler.transform(X_train)
# X_test_scaled = X_scaler.transform(X_test)
```

```python
# Create a random forest classifier.
## Check what number of estimators should be and what random state should be set to

rf_model = RandomForestClassifier(n_estimators=128, random_state=78) 
```

```python
# Fit the model

rf_model = rf_model.fit(X_train, y_train)
```

```python
# Make predictions using the testing data

predictions = rf_model.predict(X_test)
```

```python
# Create the confusion matrix and a dataframe

cm = confusion_matrix(y_test, predictions)

cm_df = pd.DataFrame(
    cm, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"])
```

```python
cm_df
```

```python
# Calculate the accuracy score

acc_score = accuracy_score(y_test, predictions)
```

```python
# Display results

print("Confusion Matrix")
display(cm_df)
print(f"Accuracy Score : {acc_score}")
print("Classification Report")
print(classification_report(y_test, predictions))
```

```python
# Calculate feature importance in the Random Forest model.
#The output from this code returns an array of scores for the features in the X_test set, whose sum equals 1.0:
importances = rf_model.feature_importances_
importances
```

```python
# Sort the features by their importance.
sorted(zip(rf_model.feature_importances_, X.columns), reverse=True)
```
