```python
import pandas as pd
import sqlite3
import csv
```

```python
con = sqlite3.connect("mushrooms.db")
cur = con.cursor()
```

```python
df = pd.read_csv('mushrooms.csv')
```

```python
df_features = df.drop(['class'], axis=1)
df_target = df[['class']]
```

```python
df_target
```

```python
sql_cols = 'create table mushroom_features (id number, '
for col in df_features.columns:
    sql_cols += col + ' varchar,\n'
    
create_features = f"{sql_cols[:-2]})"
```

```python
print(create_features)
```

```python
cur.execute(create_features)
```

```python
all_data = []
for i, row in df_features.iterrows():
    one_row = [i]
    for col in row:
        one_row.append(col)
    all_data.append(one_row)
    
all_data
```
